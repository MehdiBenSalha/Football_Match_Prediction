```python
import pandas as pd


l=["all_Turkey.csv","all_saudi.csv","all_egypt.csv"]

#concatenate all csv files in l
df=pd.concat((pd.read_csv(f) for f in l),ignore_index=True)
#drop status and xg
print(df.shape)
df.drop(['Pass Success %_away', 'Pass Success %_home', 'Goal Kicks_home',
       'Goal Kicks_away', 'Clearances Completed_away',
       'Clearances Completed_home', 'Interceptions_home',
       'Crosses Completed_home', 'Interceptions_away',
       'Crosses Completed_away'],axis=1,inplace=True)
print(df.shape)
#show null with imshow
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(),cbar=False)



```
