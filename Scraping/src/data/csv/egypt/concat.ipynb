```python
import pandas as pd

#CONCATENATE ALL CSV FILES turkey-super-lig-2013-2014.json.csv  FROM 2013-2014 TO 2023-2024
df=pd.DataFrame()
s=0
for i in range(2019,2024):
    s=s+1
    df=pd.concat([df,pd.read_csv(f"egypt-premier-league-{i}-{i+1}.json.csv")])
    print(s)
#drop status and xg
df.drop(columns=["Expected Goals (xG)_away","Expected Goals (xG)_home"],inplace=True)
print(df.shape)
#show null with imshow
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(),cbar=False)

df.to_csv("all_egypt.csv",index=False)
```

```python
def generate_commands():
    commands = []
    for year in range(2015, 2024):
        season = f"{year}-{year+1}"
        country = "turkey"
        league = f"super-lig-{season.replace('-', '-')}"
        path = f"./src/data/json/{country}"
        command = f"npm run start country={country} league={league} path={path} headless"
        commands.append(command)
    return commands

commands_list = generate_commands()
for command in commands_list:
    print(command)
```
