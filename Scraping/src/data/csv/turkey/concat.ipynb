```python
import pandas as pd

#CONCATENATE ALL CSV FILES turkey-super-lig-2013-2014.json.csv  FROM 2013-2014 TO 2023-2024
df=pd.DataFrame()
s=0
for i in range(2013,2023):
    s=s+1
    df=pd.concat([df,pd.read_csv(f"turkey-super-lig-{i}-{i+1}.json.csv")])
df=pd.concat([df,pd.read_csv(f"turkey-super-lig.json.csv")])
print(s+1)
#drop status and xg
df.drop(columns=["Expected Goals (xG)_away","Expected Goals (xG)_home"],inplace=True)
print(df.shape)
#show null with imshow
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull(),cbar=False)

df.to_csv("all_Turkey.csv",index=False)


```
